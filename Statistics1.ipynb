Question 1:

What is the difference between descriptive statistics and inferential statistics? Explain with examples.

Answer 1:

The difference between descriptive statistics and inferential statistics are:

Descriptive Statistics: Descriptive statistics are used to summarize, organize, and present data in a meaningful way. It deals with the actual data we have and helps us to understand its basic features.
For example, calculating the mean, median, mode, range, standard deviation, or creating charts and graphs from a dataset are all part of descriptive statistics. It does not make predictions or generalizations beyond the data.


Inferential Statistics: Inferential statistics are used to make predictions, decisions, or generalizations about a population based on a sample of data. It helps us to go beyond the available data using probability and hypothesis testing.
This includes methods like confidence intervals, hypothesis testing, regression analysis, and so on.
Example:
 If we take a sample of 100 customers and estimate that the average age of all customers (entire population) is 25 ± 2 years, we are using inferential statistics.

Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.

Answer 3:

Mean: Mean is the average of all data. It is calculated by adding all the values and dividing the sum by the number of values.
Example:
Let the data be: 5,1,68,20,41
So the mean of the data is (5+1+68+20+41) / 5 = 27

Median: Median is the middle value of the sorted data. We have to sort the data first to find the median. For odd number of values the median is the (n+1)/2 th value and for even data the median is the average of the two middle values, ((n/2)+ (n/2)+1) / 2.
Example1:
Data= 5,6,7,4,20,2
Sorted data: 2,4,5,6,7,20
Median =( 5+6 )/2 = 5.5
Example 2: 52, 25 , 63
Sorted data= 25,52,63
Median= 52

Mode: Mode is the value which has the highest frequency in the dataset. If all the values are present only once then the dataset will have no mode. If more than 1 value is present more than 1 time and has the same frequency then the dataset is polymodal
Example: 5,6,9,4,21,58,5,1,5
mode= 5 (highest frequency = 3)


These measures of central tendency are important to get a better idea of the data.




Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?

Answer 4:
Skewness measures the asymmetry of a distribution.


It tells us whether the data is symmetrical or leans to one side.


➤ Types:
Zero skewness → Perfectly symmetrical (like a normal distribution)


Positive skew (right-skewed) → Tail is longer on the right side; most values are on the left


Negative skew (left-skewed) → Tail is longer on the left side; most values are on the right

Kurtosis measures the "tailedness" or peakedness of a distribution.
It tells us how extreme the outliers are.


➤ Types:
Mesokurtic → Normal level of tails (like normal distribution)


Leptokurtic (High kurtosis) → Sharper peak, heavier tails → more extreme outliers


Platykurtic (Low kurtosis) → Flatter peak, lighter tails → fewer extreme outliers







```python
'''Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

(Include your Python code and output in the code box below.)


Answer 5:
'''

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
#mean
total = 0
count = 0
for num in numbers:
    total = total + num
    count = count + 1
mean = total / count

# Median
for i in range(count):
    for j in range(0, count - i - 1):
        if numbers[j] > numbers[j + 1]:
            numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]
if count % 2 == 1:
    median = numbers[count // 2]
else:
    mid1 = numbers[count // 2 - 1]
    mid2 = numbers[count // 2]
    median = (mid1 + mid2) / 2

# Mode
frequency = []
for i in range(count):
    freq = 1
    for j in range(i):
        if numbers[i] == numbers[j]:
            break
    else:
        for k in range(i + 1, count):
            if numbers[i] == numbers[k]:
                freq += 1
        frequency.append((numbers[i], freq))

max_freq = 0
for val, freq in frequency:
    if freq > max_freq:
        max_freq = freq

mode = []
for val, freq in frequency:
    if freq == max_freq:
        mode.append(val)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
'''Question 6: Compute the covariance and correlation coefficient between the
 following two datasets provided as lists in Python:

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

Answer 6:'''


list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]
n = len(list_x)

# Calculating means
sum_x = 0
sum_y = 0
for i in range(n):
    sum_x += list_x[i]
    sum_y += list_y[i]
mean_x = sum_x / n
mean_y = sum_y / n

#Calculating covariance numerator
cov_sum = 0
for i in range(n):
    cov_sum += (list_x[i] - mean_x) * (list_y[i] - mean_y)
covariance = cov_sum / n

# Calculating standard deviations
std_x_sum = 0
std_y_sum = 0
for i in range(n):
    std_x_sum += (list_x[i] - mean_x) ** 2
    std_y_sum += (list_y[i] - mean_y) ** 2
std_x = (std_x_sum / n) ** 0.5
std_y = (std_y_sum / n) ** 0.5


correlation = covariance / (std_x * std_y)

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

```python
'''Question 7: Write a Python script to draw a boxplot for the following
numeric list and identify its outliers. Explain the result:

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

Answer 7: '''


import matplotlib.pyplot as plt

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

data.sort()
n = len(data)

def find_median(lst):
    length = len(lst)
    if length % 2 == 0:
        return (lst[length // 2 - 1] + lst[length // 2]) / 2
    else:
        return lst[length // 2]

Q2 = find_median(data)
lower_half = data[:n//2]
upper_half = data[(n+1)//2:] if n % 2 != 0 else data[n//2:]

Q1 = find_median(lower_half)
Q3 = find_median(upper_half)

IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = []
for value in data:
    if value < lower_bound or value > upper_bound:
        outliers.append(value)

plt.boxplot(data, vert=False)
plt.title("Boxplot of Given Data")
plt.xlabel("Value")
plt.grid(True)
plt.show()

print("Q1:", Q1)
print("Median (Q2):", Q2)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)

```

```python
'''Question 8: You are working as a data analyst in an e-commerce company.
The marketing team wants to know if there is a relationship between advertising spend and daily sales.
 ● Explain how you would use covariance and correlation to explore this relationship.
 ● Write Python code to compute the correlation between the two lists:

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

Answer 8:

'''
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]
n = len(advertising_spend)


sum_x = 0
sum_y = 0
for i in range(n):
    sum_x += advertising_spend[i]
    sum_y += daily_sales[i]
mean_x = sum_x / n
mean_y = sum_y / n

cov = 0
for i in range(n):
    cov += (advertising_spend[i] - mean_x) * (daily_sales[i] - mean_y)
covariance = cov / n
sum_sq_x = 0
sum_sq_y = 0
for i in range(n):
    sum_sq_x += (advertising_spend[i] - mean_x) ** 2
    sum_sq_y += (daily_sales[i] - mean_y) ** 2
std_x = (sum_sq_x / n) ** 0.5
std_y = (sum_sq_y / n) ** 0.5

correlation = covariance / (std_x * std_y)
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)

 '''
 To analyze

Covariance
Tells us the direction of the relationship:


Positive → as ad spend increases, sales increase


Negative → as ad spend increases, sales decrease


But it’s not scaled, so hard to interpret strength.


➤ Correlation Coefficient (r)
Measures strength and direction of the relationship.


Range: -1 to 1


+1 = strong positive


0 = no linear relationship


-1 = strong negative


Better than covariance for comparison.
'''
```
